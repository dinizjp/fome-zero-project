```python
df = pd.read_csv('zomato_tratado.csv')
```

# Média de preço do prato para duas pessoas por país


```python
mean_price = df.groupby('country_code')['price_in_dollar'].mean().reset_index().round(2)


fig = px.bar(mean_price.sort_values('price_in_dollar', ascending=False), 
            x='country_code', 
            y='price_in_dollar',
            width= 1000,
            height= 600,
            labels={'price_in_dollar':'Preço em dólar $', 'country_code':'Países'},
            text='price_in_dollar'
)

fig.update_traces(textfont=dict(size=12),
                  textposition='outside')

colors = ['#00519e','#0d59a2','#1a62a7','#276bac','#3474b1','#417db6','#4e85bb','#5b8ec0','#6897c5',
 '#75a0c9','#82a9ce','#8fb2d3','#9cbad8','#a9c3dd','#b6cce2','#c3d5e7','#d0deec','#dde7f1']


fig.update_traces(marker_color=colors)

fig.update_layout(xaxis_title_font=dict(size=16),
                  xaxis_tickfont=dict(size=10))

fig.show()
```

# 6. Os restaurantes que aceitam pedido online são também, na média, os restaurantes que mais possuem avaliações registradas?


```python
group_mean_delivery = df.groupby('has_online_delivery')['votes'].mean().reset_index().sort_values('votes',ascending=False).round()

group_mean_delivery

fig = px.bar(group_mean_delivery, 
             x='has_online_delivery',
             y='votes',
             width=1000,
             height=600,
             labels={'has_online_delivery':'', 'votes':'Quantidade de Avaliações'},
             text='votes' 
            )


colors = ['#68AAEC', '#EC6868']


fig.update_traces(marker_color=colors)

fig.update_layout(xaxis_title_font=dict(size=16),
                  xaxis_tickfont=dict(size=10))

fig.update_traces(textfont=dict(size=12),
                  textposition='outside')
fig.update_xaxes(
        tickvals=[0, 1],
        ticktext=['Sem pedidos online', 'Com pedidos online'],
         tickfont=dict(size=16)
    )

fig.show()
```

# Os restaurantes do tipo de culinária japonesa dos Estados Unidos da América possuem um valor médio de prato para duas pessoas maior que as churrascarias americanas (BBQ)?


```python
rest_japanese = df.loc[(df['country_code'] == 'United States of America') & (df['cuisines'] == 'Japanese'), 'price_in_dollar'].mean().round(2)

rest_bbq = df.loc[(df['country_code'] == 'United States of America') & (df['cuisines'] == 'BBQ'), 'price_in_dollar'].mean().round(2)

df_result = pd.DataFrame( {'cuisines': ['Japonesa', 'BBQ'],
        'mean_price_in_dollar': [rest_japanese, rest_bbq]})

df_result



fig = px.bar(df_result, 
             x='cuisines',
             y='mean_price_in_dollar',
             width=1000,
             height=600,
             labels={'cuisines':'Nome da culinária', 'mean_price_in_dollar':'Preço médio em Dólar $'},
             text = 'mean_price_in_dollar'
            )

fig.update_traces(textfont=dict(size=12),
                  textposition='outside')

colors = ['#68AAEC', '#EC6868']
fig.update_traces(marker_color=colors)

fig.update_layout(xaxis_title_font=dict(size=16),
                  xaxis_tickfont=dict(size=14))

fig.show()
```

## Os restaurantes que fazem reservas são também, na média, os restaurantes que possuem o maior valor médio de um prato para duas pessoas?


```python
group_mean_reserv = df.groupby('has_table_booking')['price_in_dollar'].mean().reset_index().round(2)


group_mean_reserv.replace({0: 'Aceita Reserva', 1: 'Não Aceita Reserva'}, inplace=True)

fig = px.bar(group_mean_reserv, 
             x='has_table_booking',
             y='price_in_dollar',
             width=1000,
             height=600,
             labels={'has_table_booking':'Aceita reserva', 'price_in_dollar':'Preço em Dólar $'},
             text = 'price_in_dollar'
            )

fig.update_traces(textfont=dict(size=12),
                  textposition='outside')

colors = ['#68AAEC', '#EC6868']

fig.update_traces(marker_color=colors, textfont=dict(size=15))

fig.update_layout(xaxis_title_font=dict(size=16),
                  xaxis_tickfont=dict(size=14))


fig.show()
```
