# Descrição das colunas

| Column | Description |
| :----- | :---------- |
| Restaurant ID | ID do restaurante | 
| Restaurant Name | Nome do Restaurante |
| Country Code | Código do País |
| City | Nome da Cidade onde o restaurante está |
| Address | Endereço do restaurante |
| Locality | Localização e pontos de referência do restaurante |
| Locality Verbose | Localização e pontos de referência do restaurante (Mais informações) |
| Lonmgitude | Ponto geográfico de Longitude do Restaurante |
| Latitude | Ponto geográfico de Latitude do Restaurante |
| Cuisines | Tipos de Culinária servidos no restaurante |
| Average Cost for two | Preço Médio de um prato para duas pessoas no restaurante |
| Currency | Moeda do país |
| Has Table booking | Se o restaurante possui serviços de reserva; 0 - Sim; 1 - Não |
| Has Online delivery | Se o restaurante possui serviços de pedido on-line; 1 - Sim; 0 - Não |
| Is delivering now | Se o restaurante faz entregas; 0 - Sim; 1 - Não |
| Price range | Variação de preços do restaurante; 1 a 4 - Quanto maior o valor, mais caro serão os pratos |
| Aggregate rating | Nota média do restaurante |
| Rating color | Código Hexadecimal da cor do restaurante com base em sua nota média |
| Rating text | Categoria em que o restaurante está com base em sua nota média |
| Votes | Quantidade de avaliações que o restaurante já recebeu |
|Price range description| descrição dos valores em categorias : 1 = cheap, 2 = normal, 3 = expensive, 4 = gourmet |
|Price in Dollar | valores do preço médio por pessoa convertido em dolar |

# Perguntas sobre paises

## 1. Qual o nome do país que possui mais cidades registradas?

```python
df_pais_cidade = df.groupby('country_code')['city'].nunique().idxmax()

df_pais_cidade
```

## 2. Qual o nome do país que possui mais restaurantes registrados?

```python
df_pais_rest = df.groupby('country_code')['restaurant_name'].count().idxmax()

df_pais_rest
```

## 3. Qual o nome do país que possui mais restaurantes com o nível de preço igual a 4 registrados?

```python
pais_mais_gourmet = df[df['price_range'] >= 4]['country_code'].value_counts().idxmax()

pais_mais_gourmet
```

## 4. Qual o nome do país que possui a maior quantidade de tipos de culinária distintos?

```python
pais_culinaria = df.groupby('country_code')['cuisines'].nunique().idxmax()

pais_culinaria
```

## 5. Qual o nome do país que possui a maior quantidade de avaliações feitas?

```python
pais_avaliacao = df.groupby('country_code')['votes'].sum().idxmax()

pais_avaliacao
```

## 6. Qual o nome do país que possui a maior quantidade de restaurantes que fazem entrega?

```python
pais_rest_delivery = df[df['has_online_delivery'] == 1 ]['country_code'].value_counts().idxmax()

pais_rest_delivery
```

## 7. Qual o nome do país que possui a maior quantidade de restaurantes que aceitam reservas?

```python
pais_reserva = df[df['has_table_booking'] == 0 ]['country_code'].value_counts().idxmax()

pais_reserva
```

## 8. Qual o nome do país que possui, na média, a maior quantidade de avaliações registrada?

```python
media_avaliacao =  df.groupby('country_code')['votes'].mean().idxmax()
media_avaliacao
```

## 9. Qual o nome do país que possui, na média, a maior nota média registrada?


```python
media_maior = df.groupby('country_code')['aggregate_rating'].mean().idxmax()

media_maior
```

## 10. Qual o nome do país que possui, na média, a menor nota média registrada?


```python
media_menor = df.groupby('country_code')['aggregate_rating'].mean().idxmin()
media_menor
```

## 11. Qual a média de preço de um prato para dois por país?

```python
media_preco = df.groupby('country_code')['price_in_dollar'].mean().reset_index().round(2)

media_preco
```

## 6. Os restaurantes que aceitam pedido online são também, na média, os restaurantes que mais possuem avaliações registradas?


```python
group_mean_delivery = df.groupby('has_online_delivery')['votes'].mean().to_frame().reset_index().round(2)

group_mean_delivery.replace({0: 'Tem', 1: 'Não tem'})
```

```python
if group_mean_delivery.iloc[0,1 ].mean() > group_mean_delivery.iloc[1,1].mean():
    print("Os restaurantes que aceitam pedido online têm, em média, mais avaliações registradas.")
else:
    print("Os restaurantes que aceitam pedido online não têm, em média, mais avaliações registradas.")
```

## 7. Os restaurantes que fazem reservas são também, na média, os restaurantes que possuem o maior valor médio de um prato para duas pessoas?


```python
group_mean_reserv = df.groupby('has_table_booking')['price_in_dollar'].mean().to_frame().reset_index().round(2)


group_mean_reserv.replace({0: 'Tem', 1: 'Não tem'})
```

```python
if group_mean_reserv.iloc[0,1 ].mean() > group_mean_reserv.iloc[1,1].mean():
    print("Os restaurantes que aceitam reserva têm, em média, maior valor médio de um prato para dois.")
else:
    print("Os restaurantes que aceitam reserva não têm, em média, maior valor médio de um prato para dois.")
```

## 8. Os restaurantes do tipo de culinária japonesa dos Estados Unidos da América possuem um valor médio de prato para duas pessoas maior que as churrascarias americanas (BBQ)?
